```python
from matplotlib import pyplot as plt

import numpy as np
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)


plt.plot(x,y)
plt.plot(x, np.cos(x), 'r-')
plt.plot(x, -np.sin(x), 'g--')
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```
